```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load image
image = cv2.imread('WhiteHouse.jpg')

plt.title("Original Image - WhiteHouse")
plt.axis('off')
plt.imshow(image[:, :, : : -1])
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Apply Canny edge detector
canny = cv2.Canny(gray, 100, 200)

# Display output
plt.imshow(canny, cmap='gray')
plt.title('Canny Edge Detector')
plt.axis('off')
plt.show()
```
